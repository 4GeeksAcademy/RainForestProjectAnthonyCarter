```python
import pandas as pd
```

```python
train1 = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/decision-tree-project-tutorial/main/diabetes.csv")
test1 = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/decision-tree-project-tutorial/main/diabetes.csv")

train1.head()
```

```python
X_train = train1.drop(["Outcome"], axis = 1)
y_train = train1["Outcome"]
X_test = test1.drop(["Outcome"], axis = 1)
y_test = test1["Outcome"]
```

```python
pip install scikit-learn
```

```python
from sklearn.ensemble import RandomForestClassifier

model1 = RandomForestClassifier(random_state = 42)
model1.fit(X_train, y_train)
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
fig, axis = plt.subplots(2, 2, figsize = (15, 15))

tree.plot_tree(model1.estimators_[0], ax = axis[0, 0], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model1.estimators_[1], ax = axis[0, 1], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model1.estimators_[2], ax = axis[1, 0], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)
tree.plot_tree(model1.estimators_[3], ax = axis[1, 1], feature_names = list(X_train.columns), class_names = ["0", "1", "2"], filled = True)

plt.show()
```

```python
y_pred = model1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
